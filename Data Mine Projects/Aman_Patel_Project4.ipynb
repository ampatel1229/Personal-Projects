# Question 3

factor(x, levels, labels)

```r
titanic_data$Passenger_Class <- factor(titanic_data$Pclass, levels = c(1, 2, 3), labels = c("First Class", "Second Class", "Third Class"))
```

```r
titanic_data$Status <- factor(titanic_data$Survived, levels = c(0, 1), labels = c("Dead", "Alive"))
```

```r
titanic_data$Combined <- paste(titanic_data$Status, titanic_data$Sex, sep = ", ")
```

```r
combined_table = table(titanic_data$Combined)
```

```r
print(combined_table)
```

```r
barplot(combined_table, col=c("pink", "blue", "red", "maroon"), main="Survival by Sex", xlab="Status & Sex", ylab="Number of Passengers")
```
